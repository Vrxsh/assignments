```python
# Import libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, RobustScaler

# Dataset
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Task 1 — Categorical Encoding

# Gender → Label Encoding
gender_encoder = LabelEncoder()
df['Gender'] = gender_encoder.fit_transform(df['Gender'])

# City → One-Hot Encoding (drop one column to avoid dummy variable trap)
city_encoded = pd.get_dummies(df['City'], prefix='City', drop_first=True)
df = pd.concat([df.drop('City', axis=1), city_encoded], axis=1)

# Size → Label Encoding with correct order: Small=0, Medium=1, Large=2
size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)


# Task 2 — Feature Scaling

scaler = RobustScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print("\nProcessed DataFrame:\n", df)
```
